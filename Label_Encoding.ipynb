```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df_can=pd.read_csv('/content/data.csv')

df_can.head()
```

```python
df_can['diagnosis'].value_counts()
```

```python
encode1=LabelEncoder()
```

```python
encode1.fit(df_can['diagnosis'])
label=encode1.transform(df_can['diagnosis'])
```

```python
df_can['target']=label

df_can.head()
```

```python
df_can['target'].value_counts()
```

```python
df_iris=pd.read_csv('/content/iris_data.csv')
df_iris.head()
```

```python
encode2=LabelEncoder()
```

```python
encode2.fit(df_iris['Species'])

label2=encode2.transform(df_iris['Species'])
```

```python
df_iris['target']=label2

df_iris.head()
```

```python
df_iris['target'].value_counts()
```
