```python
import numpy as np
```

```python
class SigmoidPreceptron():

  def __init__(self,input_size):

    self.weights = np.random.randn(input_size)
    self.bias= np.random.randn(1)

  def sigmoid(self,x):
    return 1/(1+np.exp(-x))


  def predict(self,inputs):

    weighted_sum =np.dot(inputs,self.weights)+self.bias
    return self.sigmoid(weighted_sum)

  def fit(self , inputs , targets , learning_rate , num_epochs):

      num_examples=inputs.shape[0]

      for epoch in range(num_epochs):

        for i in range(num_examples):

          input_vectors=inputs[i]

          target=targets[i]

          prediction=self.predict(input_vectors)

          error=target-prediction

          #update weight
          gradient_weight=error*prediction*(1-prediction)*input_vectors
          self.weights+=learning_rate*gradient_weight

          #update bias
          gradient_bias=error*prediction*(1-prediction)
          self.bias+=learning_rate*gradient_bias



  def evaluate(self,inputs,targets):

    correct_predictions=0

    for input_vector, target in zip(inputs,targets):


      prediction=self.predict(input_vector)

      if prediction>=0.5 :
        prediction=1
      else:
        prediction=0

      if prediction==target:
        correct_predictions+=1

    accuracy=correct_predictions/len(inputs)
    return accuracy
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
df['Outcome'].value_counts()
```

```python
class_0=df[df['Outcome']==0]
class_1=df[df['Outcome']==1]

class_0=class_0.sample(268)
data=pd.concat([class_0,class_1])
```

```python
data.head()
```

```python
data. Outcome.value_counts()
```

```python
X=data.drop('Outcome',axis=1)
Y=data['Outcome']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled.shape[1]
```

```python
preceptron=SigmoidPreceptron(input_size=X_train_scaled.shape[1])
```

```python
Y_train = Y_train.to_numpy()
preceptron.fit(inputs=X_train, targets=Y_train, learning_rate=0.01, num_epochs=100)
```

```python
accuracy=preceptron.evaluate(X_train_scaled,Y_train)
print(f'Accuracy: {accuracy}')
```

```python
accuracy=preceptron.evaluate(X_test_scaled,Y_test)
print(f'Accuracy: {accuracy}')
```
