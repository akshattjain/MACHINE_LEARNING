```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)

print(x)
```

```python
 print(y)
```

```python
print(z)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine')
plt.title('sine wave')
plt.show()
```

```python
x=np.linspace(-10,10,20)

y=x**2

plt.plot(x,y,'g.')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```
